## Analyse data

#### 1. What is the distribution of conservation status for taxonomical categories?

```python
conserv_cat = species[species.conservation_status != "Least Concern"]\
    .groupby(["conservation_status", "category"])["scientific_name"]\
    .count()\
    .reset_index()

# print(conserv_cat.head())
conserv_cat.rename(columns={"scientific_name": "count"}, inplace=True)

stacked_data = conserv_cat.pivot(
    index='conservation_status',
    columns='category',
    values='count'
).fillna(0)

stacked_data.plot(
    kind='bar',
    stacked=True,
    figsize=(14, 8)
)
plt.show()

sns.barplot(data=conserv_cat, x="conservation_status", y="count", hue="category")
plt.show()
```

It seems that the most endangered class of species are mammals, and that there are very few species in recovery, out of which the overwhelming majority are birds. The number of threatened species is also very small, but more well-rounded (although with few/no instances of birds, reptiles, or nonvascular plants). However, the number of species of concern is extremely larger than for the other conservation statuses, with the bulk of this mainly being due to the high number of birds in this group, but also due to the large number of vascular plants and mammals (and small amount of the rest).

#### 2. Are certain types of species more likely to be endangered?

```python
# make column of boolean values: true if species is protected
species["is_protected"] = species.conservation_status != "Least Concern"

# make contingency table of taxonomical category & is_protected value
contingency_table = pd.crosstab(species.category, species.is_protected)
observed_df = contingency_table

# run chi2 contingency test to obtain expected values
chi2, p, dof, expected = chi2_contingency(contingency_table)
expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)

# prepare for side-by-side bar plots
x = np.arange(len(contingency_table.index))
width = 0.35
ax = plt.subplot()


# plot observed values
bar1 = ax.bar(x - width/2, observed_df[True], width, label="Observed")

# plot expected values
bar2 = ax.bar(x + width/2, expected_df[True], width, label="Expected")

# add labels & title
ax.set_xlabel("Category")
ax.set_ylabel("Number of Species")
ax.set_title("Observed vs. Expected Species in Protection by Taxonomical Category")
ax.set_xticks(x)
ax.set_xticklabels(observed_df.index, rotation=45, ha="right")
ax.legend()

# display the plot
plt.tight_layout()
plt.show()

# print stats
print("Chi2 value: ", chi2)
print("p value: ", p)
```

It seems that overall there is a large chi2 value and a very very small p value, meaning that the observed differences vary greatly from the expected and that this analysis is stastistically significant (high confidence).

As for the comparisons in groups, there is an enormous difference in expected and observed values for 3 groups mainly: vascular plants, which have less than half of the expected number of species in protection, and birds and mammals, which have more than triple the amount of species in protection. The other groups also have differences, but less pronounced (in terms of absolute values): amphibians, fish, and reptiles had more species in protection than expected, whereas nonvascular had fewer.

Overall, it seems all plants had fewer species in protection than expected, whereas the animals had more.

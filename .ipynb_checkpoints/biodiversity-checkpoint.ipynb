## Analyse data

#### 1. What is the distribution of conservation status for taxonomical categories?

```python
conserv_cat = species[species.conservation_status != "Least Concern"]\
    .groupby(["conservation_status", "category"])["scientific_name"]\
    .count()\
    .reset_index()

# print(conserv_cat.head())
conserv_cat.rename(columns={"scientific_name": "count"}, inplace=True)

stacked_data = conserv_cat.pivot(
    index='conservation_status',
    columns='category',
    values='count'
).fillna(0)

stacked_data.plot(
    kind='bar',
    stacked=True,
    figsize=(14, 8)
)
plt.show()

sns.barplot(data=conserv_cat, x="conservation_status", y="count", hue="category")
plt.show()
```

__Results 1:__

It seems that the most endangered class of species are mammals, and that there are very few species in recovery, out of which the overwhelming majority are birds. The number of threatened species is also very small, but more well-rounded (although with few/no instances of birds, reptiles, or nonvascular plants). However, the number of species of concern is extremely larger than for the other conservation statuses, with the bulk of this mainly being due to the high number of birds in this group, but also due to the large number of vascular plants and mammals (and small amount of the rest).

#### 2. Are certain types of species more likely to be endangered?

```python
# make column of boolean values: true if species is protected
species["is_protected"] = species.conservation_status != "Least Concern"

# make contingency table of taxonomical category & is_protected value
contingency_table = pd.crosstab(species.category, species.is_protected)
observed_df = contingency_table

# run chi2 contingency test to obtain expected values
chi2, p, dof, expected = chi2_contingency(contingency_table)
expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)

# prepare for side-by-side bar plots
x = np.arange(len(contingency_table.index))
width = 0.35
ax = plt.subplot()


# plot observed values
bar1 = ax.bar(x - width/2, observed_df[True], width, label="Observed")

# plot expected values
bar2 = ax.bar(x + width/2, expected_df[True], width, label="Expected")

# add labels & title
ax.set_xlabel("Category")
ax.set_ylabel("Number of Species")
ax.set_title("Observed vs. Expected Species in Protection by Taxonomical Category")
ax.set_xticks(x)
ax.set_xticklabels(observed_df.index, rotation=45, ha="right")
ax.legend()

# display the plot
plt.tight_layout()
plt.show()

# print stats
print("Chi2 value: ", chi2)
print("p value: ", p)
```

__Results 2:__

It seems that overall there is a large chi2 value and a very very small p value, meaning that the observed differences vary greatly from the expected and that this analysis is stastistically significant (high confidence).

As for the comparisons in groups, there is an enormous difference in expected and observed values for 3 groups mainly: vascular plants, which have less than half of the expected number of species in protection, and birds and mammals, which have more than triple the amount of species in protection. The other groups also have differences, but less pronounced (in terms of absolute values): amphibians, fish, and reptiles had more species in protection than expected, whereas nonvascular had fewer.

Overall, it seems all plants had fewer species in protection than expected, whereas the animals had more.

#### 3. Which animals are most prevalent in the parks, and what are their distribution?

```python
import string
from itertools import chain

# Function to remove punctuation from a string
def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, "")
    return text
# Apply function and separate words from string
commonNames = species[(species.category != "Vascular Plant") & (species.category != "Nonvascular Plant")]\
    .common_names\
    .apply(remove_punctuation)\
    .str.split().tolist()

# Remove duplicates
cleanRows = []
for item in commonNames:
    item = list(dict.fromkeys(item))
    cleanRows.append(item)

# Collapse into a single list
res = list(chain.from_iterable(i if isinstance(i, list) else [i] for i in cleanRows))

# Create list of occurences of words
words_counted = []
for i in res:
    x = res.count(i)
    words_counted.append((i,x))

# Create dataframe of keywords and their count
keyword_species_count = pd.DataFrame(set(words_counted), columns=["Word", "Count"]).sort_values("Count", ascending=False)

print(keyword_species_count.head(15))
```

```python
# Function to return dataframe of species which contain a word in their common names
def find_species(keyword):
    return species[species.common_names.str.contains(r"\b"+keyword+r"\b", regex=True)]

# create dataframes of observations of 5 most prevalent animals
warbler_obs = observations.merge(find_species("Warbler"), on="scientific_name")
salamander_obs = observations.merge(find_species("Salamander"), on="scientific_name")
sparrow_obs = observations.merge(find_species("Sparrow"), on="scientific_name")
snake_obs = observations.merge(find_species("Snake"), on="scientific_name")
bat_obs = observations.merge(find_species("Bat"), on="scientific_name")

# Group park observations
warbler_grouped = warbler_obs.groupby("park_name")["observations"].sum().reset_index()
salamander_grouped = salamander_obs.groupby("park_name")["observations"].sum().reset_index()
sparrow_grouped = sparrow_obs.groupby("park_name")["observations"].sum().reset_index()
snake_grouped = snake_obs.groupby("park_name")["observations"].sum().reset_index()
bat_grouped = bat_obs.groupby("park_name")["observations"].sum().reset_index()

# Create combined dataframe
combined_preval_animals = pd.concat([
    warbler_grouped.assign(Animal='Warbler'),
    salamander_grouped.assign(Animal='Salamander'),
    sparrow_grouped.assign(Animal='Sparrow'),
    snake_grouped.assign(Animal='Snake'),
    bat_grouped.assign(Animal='Bat')
])

# Plot distribution of most popular animals in parks
ax = plt.subplot()
sns.barplot(data=combined_preval_animals, x="park_name", y="observations", hue="Animal")
ax.set_xlabel("Park Name")
ax.set_ylabel("Observations")
ax.set_xticklabels(labels=combined_preval_animals.park_name, rotation=45, ha="right")
ax.legend()
plt.tight_layout()
plt.show()
```

__Results 3:__
In most parks the ranking of most observed animal species remains constant: first the Warbler, then the Salamander, then in close succession the Bat, the Snake, and the Sparrow, respectively. The exception is Yosemite National Park, where there are more observations of Snakes than of Bats. Generally, the observations of Warblers are larger than of the other species. The park with the most observations of each and all of these species is Yellowstone National Park.
